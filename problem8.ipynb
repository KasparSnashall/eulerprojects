So I decided to make it into a string as otherwise it is far too long a number for the system memory. Now we need to get the largest product of 13 adjecent numbers.So im going to split it up into groups where each group is 13 digits.

```python
groups = []
i = 0
t = thousand
while i <= 1000-13:
    groups.append(t[i]+t[i+1]+t[i+2]+t[i+3]+t[i+4]+t[i+5]+t[i+6]+t[i+7]+t[i+8]+t[i+9]+t[i+10]+t[i+11]+t[i+12])
    
    i+=1
    
```
